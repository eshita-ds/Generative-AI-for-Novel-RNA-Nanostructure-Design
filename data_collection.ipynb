```python
from datasets import load_dataset
import pandas as pd
import re

# Load the bpRNA dataset from Hugging Face
dataset = load_dataset("multimolecule/bprna", split="train")

# View one sample
sample = dataset[0]
print("Sample example:")
print(f"Sequence: {sample['sequence']}")
print(f"Dot-bracket structure: {sample['secondary_structure']}")
print(f"Annotation: {sample['structural_annotation']}")
```

```python
df = pd.DataFrame(dataset[:1000])  # Use [:1000] to work with a manageable subset

# Check DataFrame columns
print(df.columns)
df.head()
```

```python
print("Original dataset size:", len(dataset))
```

```python
import pandas as pd

# Load sample
df = pd.DataFrame(dataset[:1000])
print("Step 0: Initial shape:", df.shape)

# Drop nulls
df = df.dropna(subset=['sequence', 'secondary_structure'])
print("Step 1: After dropna:", df.shape)

# Keep only rows where sequence and secondary_structure lengths match
df = df[df['sequence'].str.len() == df['secondary_structure'].str.len()]
print("Step 2: After length match filter:", df.shape)

```

```python
# Reset index
df = df.reset_index(drop=True)

print("✅ Cleaned dataset shape:", df.shape)
df.head()
```

```python
# Save to CSV or JSON for model input
df.to_csv("bprna_preprocessed_subset.csv", index=False)
print("✅ Saved 1,000-sequence subset to bprna_preprocessed_subset.csv")
```
