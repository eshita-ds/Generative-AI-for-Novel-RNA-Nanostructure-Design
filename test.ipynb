```python
df.info()
df.describe()

print("Unique chains:", df['chain_id'].nunique())
print("Unique files:", df['file'].nunique())
print("Min residues:", df['n_residues'].min())
print("Max residues:", df['n_residues'].max())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Residue counts
sns.histplot(df['n_residues'], bins=100, log_scale=(False, True))
plt.title("Distribution of RNA chain lengths")
plt.show()

# Atom counts
sns.histplot(df['n_atoms'], bins=100, log_scale=(False, True))
plt.title("Distribution of atom counts per chain")
plt.show()

# Atoms vs residues
sns.scatterplot(x="n_residues", y="n_atoms", data=df, alpha=0.6)
plt.title("Atoms vs Residues (should be ~linear)")
plt.show()
```

```python
df['component'] = df['path'].str.extract(r'(component_\d+)')
comp_summary = df.groupby("component")[['n_residues','n_atoms']].agg(['mean','max','count'])
comp_summary.head()
```
